```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('Barley_price-1.csv')
df2 = pd.read_csv('Barley_production-1.csv')
```

```python
#Clean formatting of production values
df2['Value'] = df2['Value'].str.replace(',', '')
```

```python
#Change string to integer values
df2['Value'] = pd.to_numeric(df2['Value'])
```

```python
#Divide values by 1 million to represent them in millions
df2['Value'] = df2['Value'] / 1000000
```

```python
#rename for clarity
df2 = df2.rename(columns = {'Value':'Production (millions)'})
```

```python
prod = np.array(df2['Production (millions)'])
```

```python
#locating 1st and 99th percentile
min = np.percentile(prod, 1)
max = np.percentile(prod, 99)
```

```python
#Initiates lists of data with production values for each year
data2015 = []
data2010 = []
data2005 = []
data2000 = []
data1995 = []
data1990 = []
```

```python
#Iterates through data and separates by year
#Also only takes in values within min and max percentile ranges
for idx, yr in enumerate(df2['Year']):
    if yr == 2015 and df2['Production (millions)'][idx] > min and df2['Production (millions)'][idx] < max:
        data2015.append(df2['Production (millions)'][idx])
    if yr == 2010 and df2['Production (millions)'][idx] > min and df2['Production (millions)'][idx] < max:
        data2010.append(df2['Production (millions)'][idx])
    if yr == 2005 and df2['Production (millions)'][idx] > min and df2['Production (millions)'][idx] < max:
        data2005.append(df2['Production (millions)'][idx])
    if yr == 2000 and df2['Production (millions)'][idx] > min and df2['Production (millions)'][idx] < max:
        data2000.append(df2['Production (millions)'][idx])
    if yr == 1995 and df2['Production (millions)'][idx] > min and df2['Production (millions)'][idx] < max:
        data1995.append(df2['Production (millions)'][idx])
    if yr == 1990 and df2['Production (millions)'][idx] > min and df2['Production (millions)'][idx] < max:
        data1990.append(df2['Production (millions)'][idx])
```

```python
hist2015 = plt.hist(data2015, bins = 14, range = (0, 35), color = "purple")
plt.title("Barley Production 2015")
plt.xlabel("Production in Millions")
plt.ylabel("Frequency")
plt.ylim(0,140)

#Uncomment to dowload
#plt.savefig('barley_production_histogram_2015.pdf')

plt.show()
```
